## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
with open('Glen Powell.htm') as html_file:
    soup = BeautifulSoup(html_file)
print(soup.prettify())
soup.get_text()
```

```python
#download an image with beautifulsoup and save it in this repository
img_tags=soup.find_all('img')
img_url=img_tags[0]['src']
print(img_url)
img_url= 'https://./GlenPowell_files/wikipedia.png'
img=requests.get(img_url)
with open('image.jpg','wb') as f:
    f.write(img.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
links=soup.find_all('a')
for link in links[:10]:  
    full_link = link["href"]
    print(full_link[:100]) 
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
def clean(text):

    #convert the text to lowercase
    text = text.lower()
    #remove punctuation
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    #remove newline and space characters
    text = ' '.join(text.split())
    return text

clean(soup.text)
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
list =[]
for i, row in scripts.iterrows(): 
    link = row['link']
    if 'http' in link:
        try: 
            source=requests.get(link)
            source.raise_for_status()  # Check for HTTP errors
            try:
                soup = BeautifulSoup(source.text, "html.parser")
            except:
                continue
            text = soup.get_text()
            list.append(text[:1000])
        except requests.exceptions.RequestException as e:
            print(f"Failed to fetch {link}: {e}")
           
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
scripts['text']= list
scripts.to_csv("pudding_texts.csv", index=False)  
```

```python
API_KEY = "0fff4737afc8fff942b6872d4860af96" 
URL = f"https://api.themoviedb.org/3/movie/now_playing?api_key={API_KEY}&language=en-US&page=1"

response = requests.get(URL)
data = response.json()

movies = []
for movie in data["results"]:
    movies.append({
        "title": movie["title"],
        "release_date": movie["release_date"],
        "rating": movie["vote_average"],
        "vote_count": movie["vote_count"],
        "popularity": movie["popularity"],
        "overview": movie["overview"],
        "poster_path": movie['poster_path']
    })

df = pd.DataFrame(movies)
df
```

```python
#download the movie posters for 10 of these movies and save them to this repository
IMAGE_BASE_URL = "https://image.tmdb.org/t/p/w500"

POSTER_FOLDER = "posters"

for movie in movies:
    poster_path = movie["poster_path"]
    title = movie['title']
    
    if poster_path:
        poster_url = f"{IMAGE_BASE_URL}{poster_path}"
        
        img = requests.get(poster_url)
        
        filename = f"{title}.jpg"
        with open(filename,'wb') as f:
            f.write(img.content)    
```
